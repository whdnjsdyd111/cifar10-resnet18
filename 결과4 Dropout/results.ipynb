```python
import matplotlib.pyplot as plt
import matplotlib.image as img
import matplotlib as mpl
```

```python
test = "epoch-%d lr-%.4f batch-128 model_num3-0 acc.png" % (50, 0.1)
test = img.imread(test)
test = test[57:449, 54:578]
plt.axis('off')
plt.imshow(test)
```

```python
imgs = []

for lr in [0.1]:
    for epoch in [50, 60]:
        path = "epoch-%d lr-%.4f batch-128 model_num3-0 acc.png" % (epoch, lr)
        acc_png = img.imread(path)
        acc_png = acc_png[57:449, 45:578]
        imgs.append(acc_png)
```

```python
fig = plt.figure(dpi=1000)
rows = 3
cols = 3
i = 1

for lr in [0.1]:
    for epoch in [50, 60]:
        ax = fig.add_subplot(rows, cols, i)
        ax.imshow(imgs[i - 1])
        ax.set_title("epoch-%d lr-%.3f" % (epoch, lr))
        ax.axis("off")
        i += 1
        
# plt.subplots(constrained_layout=True)
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['font.size'] = 3
plt.subplots_adjust(wspace=-0.2, hspace=0.2)
plt.show()
```
